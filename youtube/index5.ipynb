```python
import numpy as np
from PIL import Image
```

```python
img_np = np.array(Image.open("img_1.png").convert('L'))
```

```python
img_np.shape
```

```python
Image.fromarray(img_np).show()
```

```python
kernel = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])
```

```python
k = 3
stride = 1
h_in, w_in = img_np.shape
h_out = ((h_in - k) // stride) + 1
w_out = ((w_in - k) // stride) + 1
print(h_out, w_out)
new_img_np = np.zeros((h_out, w_out))
for i in range(h_out):
    h_start = i * stride
    h_end = (i+k) * stride
    for j in range(w_out):
        w_start = j * stride
        w_end = (j+k) * stride

        region = img_np[h_start:h_end, w_start:w_end]
        new_img_np[i, j] = np.sum(region * kernel)
```

```python
Image.fromarray(new_img_np).show()
```
